```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

$1. 畫出一個 sin 函數：使用 np.linspace 來定義 x 軸的範圍，並使用 np.sin 函數計算 y 軸的值，最後使用 plt.plot 來畫出函數圖形。$


```python
plt.figure(figsize=(5, 5))
x = np.linspace(-10, 10, 10000)
y = np.sin(x)
plt.plot(x, y)
```

$2. 畫出一個 cos 函數：跟上面同樣的方法，只需將 np.sin 改為 np.cos 即可。$

```python
plt.figure(figsize=(5, 5))
x = np.linspace(-10, 10, 10000)
y = np.cos(x)
plt.plot(x, y)
```

$3. 畫出一個 x^2 的圖形：使用 np.linspace 來定義 x 軸的範圍，並使用平方運算符 ** 計算 y 軸的值，最後使用 plt.plot 來畫出函數圖形。$

```python
plt.figure(figsize=(5, 5))
x = np.linspace(-10, 10, 10000)
y = x**2
plt.plot(x, y)
```

$4. 畫出一個 \frac{1}{x} 的圖形：使用 np.linspace 來定義 x 軸的範圍，並使用除法運算符 / 計算 y 軸的值，最後使用 plt.plot 來畫出函數圖形。注意需要特判 x=0 的情況，因為無限大不能被計算。
$

```python
plt.figure(figsize=(5, 5))
x1 = np.linspace(-10, 0, 5000)
x2 = np.linspace(0, 10, 5000)
y1 = 1/x1
y2 = 1/x2
plt.plot(x1, y1)
plt.plot(x2, y2)
```

```python
x = np.linspace(-1, 1, 10000)
y1 = np.sqrt(1 - x**2)
y2 = -np.sqrt(1 - x**2)

plt.figure(figsize=(5, 5)) #將大小固定不然會看起來很像橢圓
plt.plot(x, y1, label='y = sqrt(x^3 - x^4)')
plt.plot(x, y2, label='y = -sqrt(x^3 - x^4)')
plt.title('x^2 + y^2 = 1')
plt.legend()
plt.plot(x, y1, color='red', linewidth=5)
plt.plot(x, y2, color='blue', linewidth=5)
```
